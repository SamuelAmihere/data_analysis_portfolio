### * Analysis on Market based on profit and sales

```python

# Market information (Average and Total). Note that it is sorted according to profit

sales_profit_mkt = data_copy_markets[['Sales', 'Profit']]

def Market_info_avg(df):
    
    market_means = df.groupby('Market').mean() #mean count sales in every market
        
    market_means_sort = market_means.sort_values('Profit',ascending=False)  # Sort market according to mean sales


    return print(market_means_sort)

  
    
def Market_info_sum(df):
    
        
    market_sums = df.groupby('Market').sum() #count profit in
                                                                          #every market
    
    market_sums_sort = market_sums.sort_values('Profit',ascending=False)  # Sort market according to mean sales
    

    return print(market_sums_sort)   

```

```python
Market_info_avg(sales_profit_mkt)
```

```python
Market_info_sum(sales_profit_mkt)
```

**1. Determining the correlation between discount and sales in each market**


```python
# 1. Determining the correlation between discount and sales in each market

data_APAC = data_mkt.loc[['APAC']][['Discount', 'Sales']].corr()
data_Africa = data_mkt.loc[['Africa']][['Discount', 'Sales']].corr()
data_Canada = data_mkt.loc[['Canada']][['Discount', 'Sales']].corr()
data_EMEA = data_mkt.loc[['EMEA']][['Discount', 'Sales']].corr()
data_EU = data_mkt.loc[['EU']][['Discount', 'Sales']].corr()
data_LATAM = data_mkt.loc[['LATAM']][['Discount', 'Sales']].corr()
data_US = data_mkt.loc[['US']][['Discount', 'Sales']].corr()


#Inspecting Discount and Sales in every Market

prof_sal_markt_1 = pd.concat([data_mkt.loc[['APAC']],data_mkt.loc[['Africa']],
                            data_mkt.loc[['Canada']],data_mkt.loc[['EMEA']],
                            data_mkt.loc[['EU']],data_mkt.loc[['LATAM']],
                            data_mkt.loc[['US']]],keys=['APAC','Africa','Canada','EMEA','EU',
               'LATAM','US'])

prof_sal_markt_inspect = pd.concat([data_mkt.loc[['APAC']].max(),data_mkt.loc[['Africa']].max(),
                            data_mkt.loc[['Canada']].max(),data_mkt.loc[['EMEA']].max(),
                            data_mkt.loc[['EU']].max(),data_mkt.loc[['LATAM']].max(),
                            data_mkt.loc[['US']].max()],keys=['APAC','Africa','Canada','EMEA','EU',
               'LATAM','US'])

# Generating the correlation 
corr_market = pd.concat([data_APAC,data_Africa,data_Canada,data_EMEA,data_EU,
               data_LATAM,data_US],keys=['APAC','Africa','Canada','EMEA','EU',
               'LATAM','US'])
```
